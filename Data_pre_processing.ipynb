```python
data=pd.read_csv('/content/drive/My Drive/texts.csv',  lineterminator='\n')
```

```python
texts = data.text
```

```python
filter_words = []
Stopwords = stopwords.words('arabic')
filter_words = [word for word in texts if word not in Stopwords]
```

**Removing punctuation & Emoji**


```python
texts = []
```

```python
emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticons
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
    "]+", flags=re.UNICODE)

english_pattern = r'[a-zA-Z]'
numbers_pattern = r'[0-9]'
newLine_pattern = r'\n'
noise = re.compile(""" ّ    | # Tashdid
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
```

```python
for text in filter_words:
  text = re.sub(r'[?!.;:,،؟_#@/-]', '', text)
  text = emoji_pattern.sub(r'', text)
  text = re.sub(english_pattern, '', text)
  text = re.sub(numbers_pattern, '', text)
  text = noise.sub(r'', text)
  text = re.sub(newLine_pattern, '', text)
  texts.append(text)
```

```python
dataFrame = pd.DataFrame({'preprocessed_text': texts, 'dialect': data.dialect})
```
